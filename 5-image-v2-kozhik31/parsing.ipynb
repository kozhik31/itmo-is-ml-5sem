```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
template_path = '/content/drive/MyDrive/Colab_Notebooks/templates/'
base_url = 'https://www.iamcook.ru/event/everyday/'
ans = requests.get('https://www.iamcook.ru/event/everyday/1')
for i in range(1, 201):
    url = base_url + str(i)
    ans = requests.get(url)
    path = template_path + str(i) + '.html'
    with open(path, 'w') as file:
        file.write(ans.text)

    if i%10 == 0:
        print(i)
```

```python
pic_path = '/content/drive/MyDrive/Colab_Notebooks/pictures/'
data_path = '/content/drive/MyDrive/Colab_Notebooks/data.csv'
template_path = '/content/drive/MyDrive/Colab_Notebooks/templates/'
RE_KKAL = re.compile(r"(\d+)\s*")
data = []
columns = ['img_path', 'kkal']

session = requests.Session()
def load_pic(url, path):
    resp = session.get(url)
    with open(path, 'wb') as f:
        f.write(resp.content)
```

```python
id = 0
for i in range(1, 201):
    path = template_path + str(i) + '.html'
    print(i)
    with open(path, 'r') as file:
        soup = BeautifulSoup(file.read(), 'html.parser')
        meals = soup.find_all('div', class_='recblockwide')
        for meal in meals:
            id += 1
            img = meal.find('img')
            kkal_div = meal.find('div', class_='energy tt')
            if img is None or kkal_div is None:
                continue

            img_url = img['src']
            img_path = pic_path + str(id) + '.jpg'
            load_pic(img_url, img_path)

            kkal = int(RE_KKAL.search(kkal_div.text).group(1))
            data.append([img_path, kkal])



df = pd.DataFrame(data, columns=columns)
df.to_csv(data_path, index=False, encoding="utf-8")
```
